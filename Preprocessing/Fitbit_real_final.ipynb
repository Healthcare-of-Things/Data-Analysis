```python
# 개인 폴더에서 필요한 파일 path load

folder = r"C:\Users\ballj\OneDrive\바탕 화면\Fitbit_JM"

things_path = glob.glob(f"{folder}/Active Zone Minutes (AZM)/*")
things_path.extend(glob.glob(f"{folder}/Sleep Score/*"))
things_path.extend(glob.glob(f"{folder}/Stress Journal/CEDA*"))
things_path.extend(glob.glob(f"{folder}/Temperature/Wrist Temperature - *"))

WT_count = len(glob.glob(f"{folder}/Temperature/Wrist Temperature - *"))
```

```python
# 필요한 columns

AZM_col = ['date_time', 'total_minutes']
sleep_col = ['timestamp', 'deep_sleep_in_minutes']
stress_col = ['timestamp', 'eda_level_real']
temp_col = ['recorded_time', 'temperature']

# 파일 수가 변동적인 temperature만 개수 counting
things = ['AZM', 'sleep', 'stress']
for i in range(WT_count):
    things.append('temp')

things_col = [AZM_col, sleep_col, stress_col]
for i in range(WT_count):
    things_col.append(temp_col)
    
things_col
```

```python
# 파일 읽어와서 timezone 변환 (pd.datetime)

def read_filtered_csv(name, path, col):
    name = pd.read_csv(path)
    name = name[col]
    name[col[0]] = pd.to_datetime(name[col[0]])
    return name
```

```python
# File Load

for i in range(len(things_path)):
    things[i] = read_filtered_csv(things[i], things_path[i], things_col[i])
```

### AZM

```python
azm = things[0]
azm = azm.rename(columns={'date_time':'timestamp', 'total_minutes':'BM'})

azm['timestamp']
```

```python
seq1 = []
seq2 = []

# for x1 in azm['timestamp']:
#     seq1.append(x1 - timedelta(seconds = 20))
#     seq1.append(x1)
#     seq1.append(x1 + timedelta(seconds = 20))
    
# for x1 in azm['BM']:
#     seq2.append(x1)
#     seq2.append(x1)
#     seq2.append(x1)
    
time_interval = 20
# 원본 데이터의 시간 간격은 1분
original_interval = 60
num_timestamps = original_interval // time_interval # 20초로 하면 3, 15초로 하면 4
half_point = num_timestamps // 2 # 20초로 하면 1, 15초로 하면 2

for x1 in azm['timestamp']:
    if half_point % 2 != 0:
        start_timestamp = x1 - timedelta(seconds=(half_point * time_interval))
        # 20초로 하면, 15:57:00 - 1*20 : 15:56:40
        for n in range(num_timestamps): # 0,1,2
            new_timestamp = start_timestamp + timedelta(seconds = (n*time_interval))
            seq1.append(new_timestamp)

    else:
        start_timestamp = x1 - timedelta(seconds=((half_point-1) * time_interval))
        # 15초로 하면, 15:57:00 - 1*15 : 15:56:45
        for n in range(num_timestamps):# 0,1,2,3
            new_timestamp = start_timestamp + timedelta(seconds = (n*time_interval))
            seq1.append(new_timestamp)

for x1 in azm['BM']:
    for _ in range(num_timestamps):
        seq2.append(x1)

azm_list = {'timestamp': seq1, 'BM': seq2}
azm_final = pd.DataFrame(azm_list)

azm_final
```

#### 최근접 이웃 보간법 VS 양선형 보간법

```python
# 최근접

seq5 = []
seq6 = []

for x1 in eda['timestamp']:
    seq5.append(x1 - timedelta(seconds = 20))
    seq5.append(x1)
    seq5.append(x1 + timedelta(seconds = 20))
    
for x1 in eda['eda_level_real']:
    seq6.append(x1)
    seq6.append(x1)
    seq6.append(x1)
```

```python
eda_list_1 = {'timestamp': seq5, 'eda': seq6}
eda_final_1 = pd.DataFrame(eda_list_1)
eda_final_1
```

```python
# 양선형

seq = []
seq0 = []

for i in range(1,len(eda['timestamp'])-1):
    seq.append(eda.iloc[i,0] - timedelta(seconds = 20))
    seq0.append(round(eda.iloc[i,1]-(eda.iloc[i,1]-eda.iloc[i-1,1])/3,2))
    seq.append(eda.iloc[i,0])
    seq0.append(round(eda.iloc[i,1],2))
    seq.append(eda.iloc[i,0] + timedelta(seconds = 20))
    seq0.append(round(eda.iloc[i,1]+(eda.iloc[i+1,1]-eda.iloc[i,1])/3,2))
```

```python
eda_list_2 = {'timestamp': seq, 'eda': seq0}
eda_final_2 = pd.DataFrame(eda_list_2)
eda_final_2
```
