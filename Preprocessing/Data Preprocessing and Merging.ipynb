#### Data parsing function

```python
#리스트 값을 row로 변환하는 함수
def parse_raw_data(dataframe, col_name):
    col_str = dataframe.iloc[0][col_name]
    col_str = col_str.strip('[]')
    col_list = [float(val) for val in col_str.split(',')]  # 쉼표로 구분된 값을 리스트로 변환
    col_data = pd.DataFrame({col_name: col_list})
    return col_data
```

#### Experiment time function

```python
#측정 시간 출력 함수
from datetime import datetime, timedelta

def time_difference(dataframe, start_time_col, finish_time_col):
    start_time = dataframe.iloc[0][start_time_col]
    start_time = datetime.strptime(start_time, '%Y-%m-%d %H:%M:%S')
    
    finish_time = dataframe.iloc[0][finish_time_col]
    finish_time = datetime.strptime(finish_time, '%Y-%m-%d %H:%M:%S')
    
    # 두 datetime 객체 간의 차이 계산
    time_difference = finish_time - start_time
    time_difference = time_difference.total_seconds()
    
    return time_difference
```

**실험 초기 인식 오류 기간 비교 후, 삭제할 부분 삭제**

```python
def count_initial_same_values(series):
    initial_value = series.iloc[0]
    count = 0
    for value in series:
        if value == initial_value:
            count += 1
        else:
            break
    return count
```

```python
def process_start_time_trash_sec(dt):
    # 입력된 시간에서 분은 1 더하고 초는 버림
    rounded_time = dt + timedelta(minutes=1) - timedelta(seconds=dt.second)
        
    # 두 시간 간의 초 차이 계산
    time_difference = (rounded_time - dt).total_seconds()
    
    # 20초로 나눈 나머지 계산 ex. 12
    remainder = time_difference % 20.0
    
    # 나머지가 15초 미만의 데이터는 버리고, 15초 이상의 데이터는 버리는 것 없이 그대로 사용하기
    # 이러면 절사평균할 때 앞에 살릴 것들은 따로 절사평균 진행
    if 15 <= remainder:
        return False
    
    else:
        return remainder
    
# ex. 2023-10-15 00:21:04 -> remainder : 16 (False)
# ex. 2023-10-15 00:21:15 -> remainder : 5
```

```python
remainder = process_start_time_trash_sec(merged_df.index[0])

def nearest_time_rounding(dt):
    # 주어진 시간의 초를 추출
    seconds = dt.second
    # 0, 20, 40초 중 가장 가까운 값을 찾기 위한 리스트
    time_points = [0, 20, 40]
    # 가장 가까운 초 값 찾기
    nearest = min(time_points, key=lambda x: abs(x - seconds))
    # 찾은 초 값으로 시간 변경
    rounded_time = dt.replace(second=nearest, microsecond=0)
    
    return rounded_time

# ex. 2023-10-15 00:21:04 -> remainder : 16 (False)
if remainder == False:
    # 첫 번째 인덱스를 가장 가까운 0, 20, 40초 중 하나로 변경 -> 그 시간부터 20초 단위로 절사평균 진행하면 됨
    start_time = nearest_time_rounding(merged_df.index[0])
    
    # 인덱스를 리스트로 변환하고 첫 번째 값을 수정
    new_index = merged_df.index.tolist()
    new_index[0] = start_time
    merged_df.index = new_index
    
# ex. 2023-10-15 00:21:15 -> remainder : 5
else:
    # remainder가 15초 미만이면 버리고 그 이후부터 시작
    cutting_time = timedelta(seconds=remainder)
    merged_df = merged_df[merged_df.index >= merged_df.index[0] + cutting_time]
```

```python
remainder_other = process_start_time_trash_sec(merged_df_other.index[0])
if remainder_other == False:
    # 첫 번째 인덱스를 가장 가까운 0, 20, 40초 중 하나로 변경 -> 그 시간부터 20초 단위로 절사평균 진행하면 됨
    start_time_other = nearest_time_rounding(merged_df_other.index[0])
    # 인덱스를 리스트로 변환하고 첫 번째 값을 수정
    new_index_other = merged_df_other.index.tolist()
    new_index_other[0] = start_time_other
    merged_df_other.index = new_index_other

else:
    # remainder가 15초 미만이면 버리고 그 이후부터 시작
    cutting_time_other = timedelta(seconds=remainder_other)
    merged_df_other = merged_df_other[merged_df_other.index >= merged_df_other.index[0] + cutting_time_other]
```
